```python
import numpy as np
import pandas as pd
from sklearn.cluster import *

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import pairwise_distances_argmin
```

```python
df = pd.read_excel('./data_clustering.xlsx')
df = df.dropna(axis=0)
X = df.values

X = StandardScaler().fit_transform(X)
```

```python
k_means = KMeans(init='k-means++', n_clusters=4, n_init=10).fit(X)
k_means_cluster_centers = [[-1, 1], [1, -1], [-1, -1], [1, 1]]
k_means_labels = pairwise_distances_argmin(X, k_means_cluster_centers)
k_order = np.argsort(k_means.cluster_centers_[:, 0] + k_means.cluster_centers_[:, 1]*0.1)

plt.scatter(df["x"], df["y"], c=k_means_labels, s=10)

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

plt.show()
```
