```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

data = pd.read_excel('./data_clustering.xlsx')
data = data.dropna(axis=0)
A = data[['PSPscore_w0', 'SWN_Total']].values

X = StandardScaler().fit_transform(A)
clustering = AgglomerativeClustering(n_clusters=4, linkage='complete').fit_predict(X)
pd.DataFrame(clustering).to_excel("./hier_clustering.xlsx")

plt.figure(figsize=(8, 6))
plt.scatter(A[:, 0], A[:, 1], c=clustering, cmap='Set2', s=30)
plt.xlabel('PSP')
plt.ylabel('SWN-K')

plt.show()
```
