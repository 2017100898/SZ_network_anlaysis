```python
import os.path as op

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from bct.algorithms.centrality import *
from bct.algorithms.core import *
from bct.algorithms.modularity import *
from bct.algorithms.distance import *
from bct.algorithms.degree import *
from bct.algorithms.physical_connectivity import *
import networkx

from numpy import inf
from scipy import stats
```

```python
# input data : correlation network matrix

data0 = pd.read_csv('./Desktop/knet00.csv', index_col='Unnamed: 0')
N0 = data0.values
Name0 = data0.columns
a0 = np.reciprocal(N0)
a0[a0 == inf] = 0
a0 = np.absolute(a0)

data1 = pd.read_csv('./Desktop/knet01.csv', index_col='Unnamed: 0')
N1 = data1.values
Name1 = data1.columns
a1 = np.reciprocal(N1)
a1[a1 == inf] = 0
a1 = np.absolute(a1)

data2 = pd.read_csv('./Desktop/knet02.csv', index_col='Unnamed: 0')
N2 = data2.values
Name2 = data2.columns
a2 = np.reciprocal(N2)
a2[a2 == inf] = 0
a2 = np.absolute(a2)

data3 = pd.read_csv('./Desktop/knet03.csv', index_col='Unnamed: 0')
N3 = data3.values
Name3 = data3.columns
a3 = np.reciprocal(N3)
a3[a3 == inf] = 0
a3 = np.absolute(a3)

data4 = pd.read_csv('./Desktop/knet04.csv', index_col='Unnamed: 0')
N4 = data4.values
Name4 = data4.columns
a4 = np.reciprocal(N4)
a4[a4 == inf] = 0
a4 = np.absolute(a4)
```

```python
#Strength

result0 = strengths_und(N0)
strength0 = stats.zscore(result0)
result1 = strengths_und(N1)
strength1 = stats.zscore(result1)
result2 = strengths_und(N2)
strength2 = stats.zscore(result2)
result3 = strengths_und(N3)
strength3 = stats.zscore(result3)
result4 = strengths_und(N4)
strength4 = stats.zscore(result4)

fig= plt.figure(figsize=(3,7))
plt.plot(strength0, Name0,  color='lightcoral', marker='o', linewidth = 1.0, label = "all")
plt.plot(strength1, Name1,  color='gold', marker='o', linewidth = 1.0,label = "cluster1")
plt.plot(strength2, Name2,  color='skyblue', marker='o', linewidth = 1.0,label = "cluster2")
plt.plot(strength3, Name3,  color='mediumpurple', marker='o', linewidth = 1.0,label = "cluster3")
plt.plot(strength4, Name4,  color='lightseagreen', marker='o', linewidth = 1.0,label = "cluster4")

plt.title('Strength', fontsize=14)
plt.grid(True)
ax = plt.subplot(111)

# To hide the exposure of detailed results.
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
#Betweenness

result0 = betweenness_wei(N0)
result0 = result0//2
bet0 = stats.zscore(result0)

result1 = betweenness_wei(a1)
result1 = result1//2
bet1 = stats.zscore(result1)

result2 = betweenness_wei(a2)
result2 = result2//2
bet2 = stats.zscore(result2)

result3 = betweenness_wei(a3)
result3 = result3//2
bet3 = stats.zscore(result3)

result4 = betweenness_wei(a4)
result4 = result4//2
bet4 = stats.zscore(result4)

fig= plt.figure(figsize=(3,7))
plt.plot(bet0, Name0,  color='lightcoral', marker='o', linewidth = 1.0, label = "all")
plt.plot(bet1, Name1,  color='gold', marker='o', linewidth = 1.0,label = "cluster1")
plt.plot(bet2, Name2,  color='skyblue', marker='o', linewidth = 1.0,label = "cluster2")
plt.plot(bet3, Name3,  color='mediumpurple', marker='o', linewidth = 1.0,label = "cluster3")
plt.plot(bet4, Name4,  color='lightseagreen', marker='o', linewidth = 1.0,label = "cluster4")

plt.title('Betweenness', fontsize=14)
plt.grid(True)
ax = plt.subplot(111)


# To hide the exposure of detailed results.
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
# Closeness

res0 = []
result0 = distance_wei(a0)
for i in range (len(result0[0])):
    res0.append(np.sum(result0[0][i]))
res0 = np.reciprocal(res0)
close0 = stats.zscore(res0)

res1 = []
result1 = distance_wei(a1)
for i in range (len(result1[0])):
    res1.append(np.sum(result1[0][i]))
res1 = np.reciprocal(res1)
close1 = stats.zscore(res1)

res2 = []
result2 =distance_wei(a2)
for i in range (len(result2[0])):
    res2.append(np.sum(result2[0][i]))
res2 = np.reciprocal(res2)
close2 = stats.zscore(res2)

res3 = []
result3 =distance_wei(a3)
for i in range (len(result3[0])):
    res3.append(np.sum(result3[0][i]))
res3 = np.reciprocal(res3)
close3 = stats.zscore(res3)

res4 = []
result4 =distance_wei(a4)
for i in range (len(result4[0])):
    res4.append(np.sum(result4[0][i]))
res4 = np.reciprocal(res4)
close4 = stats.zscore(res4)


fig= plt.figure(figsize=(3,7))
plt.plot(close0, Name0,  color='lightcoral', marker='o', linewidth = 1.0, label = "all")
plt.plot(close1, Name1,  color='gold', marker='o', linewidth = 1.0,label = "cluster1")
plt.plot(close2, Name2,  color='skyblue', marker='o', linewidth = 1.0,label = "cluster2")
plt.plot(close3, Name3,  color='mediumpurple', marker='o', linewidth = 1.0,label = "cluster3")
plt.plot(close4, Name4,  color='lightseagreen', marker='o', linewidth = 1.0,label = "cluster4")

plt.title('Closeness', fontsize=14)
plt.grid(True)
ax = plt.subplot(111)

# To hide the exposure of detailed results.
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
